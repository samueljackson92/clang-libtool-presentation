```python
%matplotlib inline
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
with open("/Users/samueljackson/git/clang-stuff/output") as f:
    data = json.load(f)
```

```python
s = pd.Series({d: data[d]['length'] for d in data})
print s.describe()
ax = s.plot('hist')
ax.set_xlabel("Length of Method")
ax.spines['bottom'].set_color('#ffffff')
ax.spines['top'].set_color('#ffffff') 
ax.spines['right'].set_color('#ffffff')
ax.spines['left'].set_color('#ffffff')

ax.tick_params(axis='x', colors='#ffffff')
ax.tick_params(axis='y', colors='#ffffff')

ax.yaxis.label.set_color('#ffffff')
ax.xaxis.label.set_color('#ffffff')

ax.title.set_color('#ffffff')
```

```python
top_meth_len = s.sort_values(ascending=False)[:20]
ax = top_meth_len.plot('barh')
ax.set_xlabel("Length of Method")
ax.spines['bottom'].set_color('#ffffff')
ax.spines['top'].set_color('#ffffff') 
ax.spines['right'].set_color('#ffffff')
ax.spines['left'].set_color('#ffffff')

ax.tick_params(axis='x', colors='#ffffff')
ax.tick_params(axis='y', colors='#ffffff')

ax.yaxis.label.set_color('#ffffff')
ax.xaxis.label.set_color('#ffffff')

ax.title.set_color('#ffffff')
```

```python
depth = pd.Series({d: data[d]['depth'] for d in data})
print depth.describe()
ax = depth.plot('hist')
ax.set_xlabel("Depth of Method")
ax.spines['bottom'].set_color('#ffffff')
ax.spines['top'].set_color('#ffffff') 
ax.spines['right'].set_color('#ffffff')
ax.spines['left'].set_color('#ffffff')

ax.tick_params(axis='x', colors='#ffffff')
ax.tick_params(axis='y', colors='#ffffff')

ax.yaxis.label.set_color('#ffffff')
ax.xaxis.label.set_color('#ffffff')

ax.title.set_color('#ffffff')
```

```python
top_meth_depth = depth.sort_values(ascending=False)[:20]
ax = top_meth_depth.plot('barh')

ax.set_xlabel("Depth of Method")
ax.spines['bottom'].set_color('#ffffff')
ax.spines['top'].set_color('#ffffff') 
ax.spines['right'].set_color('#ffffff')
ax.spines['left'].set_color('#ffffff')

ax.tick_params(axis='x', colors='#ffffff')
ax.tick_params(axis='y', colors='#ffffff')

ax.yaxis.label.set_color('#ffffff')
ax.xaxis.label.set_color('#ffffff')

ax.title.set_color('#ffffff')
```

```python
complexity = pd.Series({d: data[d]['complexity'] for d in data})
print complexity.describe()


print"1 <= x <= 10 {:.2f}%".format(float(complexity[(complexity >= 1) & (complexity<=10)].size) / complexity.size *100)
print"11 <= x <= 20 {:.2f}%".format(float(complexity[(complexity >= 11) & (complexity<=20)].size) / complexity.size *100)
print"21 <= x <= 50 {:.2f}%".format(float(complexity[(complexity >= 21) & (complexity<=50)].size) / complexity.size *100)
print"x > 50 {:.2f}%".format(float(complexity[complexity>50].size) / complexity.size *100)

ax = complexity.plot('hist', bins=100)
ax.set_xlabel("Complexity of Method")
ax.spines['bottom'].set_color('#ffffff')
ax.spines['top'].set_color('#ffffff') 
ax.spines['right'].set_color('#ffffff')
ax.spines['left'].set_color('#ffffff')

ax.tick_params(axis='x', colors='#ffffff')
ax.tick_params(axis='y', colors='#ffffff')

ax.yaxis.label.set_color('#ffffff')
ax.xaxis.label.set_color('#ffffff')

ax.title.set_color('#ffffff')
```
